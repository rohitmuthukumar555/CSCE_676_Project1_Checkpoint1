1. **Strong course alignment:** Graph mining (centrality, PageRank, community detection) is covered in HW3–HW4; this benchmark provides standardized graphs to apply and extend those techniques.

## D1. Setup and load graph data

```python
import os
import pandas as pd
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set_style('whitegrid')

DATA_DIR = os.path.join('.', 'data')

# Choose one: 'raw_Small' (has edges + labels + sens), 'cora', or 'credit'
DATASET_NAME = 'raw_Small'

def load_graph_fairness_data(data_dir, name):
    """Load graph + labels + sensitive attribute from Graph-Mining-Fairness-Data format."""
    G = nx.DiGraph()
    if name == 'raw_Small':
        base = os.path.join(data_dir, 'raw_Small')
        edges_path = os.path.join(base, 'edgelist_Small.txt')
        labels_path = os.path.join(base, 'labels_Small.txt')
        sens_path = os.path.join(base, 'sens_Small.txt')
        if not os.path.exists(edges_path):
            return None
        # Edges: src \t dst
        with open(edges_path) as f:
            for line in f:
                line = line.strip()
                if not line:
                    continue
                parts = line.split()
                if len(parts) >= 2:
                    G.add_edge(parts[0], parts[1])
        # Labels: node_id \t label
        labels = {}
        with open(labels_path) as f:
            for line in f:
                parts = line.strip().split()
                if len(parts) >= 2:
                    labels[parts[0]] = float(parts[1])
        # Sensitive: node_id \t sens
        sens = {}
        with open(sens_path) as f:
            for line in f:
                parts = line.strip().split()
                if len(parts) >= 2:
                    sens[parts[0]] = float(parts[1])
        for n in G.nodes():
            G.nodes[n]['label'] = labels.get(n, np.nan)
            G.nodes[n]['sensitive'] = sens.get(n, np.nan)
        return G

    if name == 'cora':
        base = os.path.join(data_dir, 'cora')
        cites_path = os.path.join(base, 'cora.cites')
        content_path = os.path.join(base, 'cora.content')
        if not os.path.exists(cites_path) or not os.path.exists(content_path):
            return None
        # cora.content: node_id \t feat1 \t ... \t label (tab-separated)
        content = pd.read_csv(content_path, sep='\t', header=None)
        node_ids = content[0].astype(str)
        label_col = content.columns[-1]
        labels = dict(zip(node_ids, content[label_col]))
        for _, row in content.iterrows():
            n = str(row[0])
            G.add_node(n, label=row[label_col], sensitive=np.nan)  # no sens in cora
        with open(cites_path) as f:
            for line in f:
                parts = line.strip().split()
                if len(parts) >= 2:
                    G.add_edge(parts[0], parts[1])
        return G

    if name == 'credit':
        base = os.path.join(data_dir, 'credit')
        edges_path = os.path.join(base, 'credit_edges.txt')
        csv_path = os.path.join(base, 'credit.csv')
        if not os.path.exists(edges_path) or not os.path.exists(csv_path):
            return None
        nodes_df = pd.read_csv(csv_path)
        n_nodes = len(nodes_df)
        # Node id = row index (string)
        for i in range(n_nodes):
            G.add_node(str(i))
        with open(edges_path) as f:
            for line in f:
                parts = line.strip().split()
                if len(parts) >= 2:
                    u, v = int(float(parts[0])), int(float(parts[1]))
                    if 0 <= u < n_nodes and 0 <= v < n_nodes:
                        G.add_edge(str(u), str(v))
        # Label: NoDefaultNextMonth (0/1), Sensitive: Age (binarize e.g. Age>=25 -> 1)
        label_col = 'NoDefaultNextMonth' if 'NoDefaultNextMonth' in nodes_df.columns else nodes_df.columns[0]
        sens_col = 'Age' if 'Age' in nodes_df.columns else nodes_df.columns[2]
        for i in range(n_nodes):
            n = str(i)
            G.nodes[n]['label'] = int(nodes_df.iloc[i][label_col])
            age = nodes_df.iloc[i][sens_col]
            G.nodes[n]['sensitive'] = 1 if age > 25 else 0
        return G

    return None

    
G = load_graph_fairness_data(DATA_DIR, DATASET_NAME)
if G is None:
    print(f'Dataset "{DATASET_NAME}" not found. Using synthetic karate graph.')
    G = nx.karate_club_graph().to_directed()
    np.random.seed(42)
    for i, n in enumerate(G.nodes()):
        G.nodes[n]['sensitive'] = np.random.choice([0, 1])
        G.nodes[n]['label'] = np.random.choice([0, 1])
else:
    print(f'Loaded dataset: {DATASET_NAME} (Graph-Mining-Fairness-Data)')

print('Nodes:', G.number_of_nodes(), 'Edges:', G.number_of_edges())
print('Directed:', G.is_directed())
```

## D2. Graph structure: degree distribution and density

```python
degrees = [d for _, d in G.degree()]
print('Degree - min, max, mean, median:', min(degrees), max(degrees), np.mean(degrees), np.median(degrees))
n = G.number_of_nodes()
m = G.number_of_edges()
density = 2 * m / (n * (n - 1)) if not G.is_directed() and n > 1 else m / (n * (n - 1)) if n > 1 else 0
print('Density:', round(density, 6))

plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.hist(degrees, bins=20, edgecolor='black', alpha=0.7)
plt.xlabel('Degree')
plt.ylabel('Count')
plt.title('Degree distribution')
plt.subplot(1, 2, 2)
plt.boxplot(degrees)
plt.ylabel('Degree')
plt.title('Degree (box)')
plt.tight_layout()
plt.show()
```

## D3. Connectivity: connected components

```python
if G.is_directed():
    wccs = list(nx.weakly_connected_components(G))
    sccs = list(nx.strongly_connected_components(G))
    print('Weakly connected components:', len(wccs))
    print('Strongly connected components:', len(sccs))
    sizes_wcc = sorted([len(c) for c in wccs], reverse=True)
    sizes_scc = sorted([len(c) for c in sccs], reverse=True)
    print('Largest WCC size:', sizes_wcc[0] if sizes_wcc else 0)
    print('Largest SCC size:', sizes_scc[0] if sizes_scc else 0)
else:
    comps = list(nx.connected_components(G))
    print('Connected components:', len(comps))
    sizes = sorted([len(c) for c in comps], reverse=True)
    print('Largest component size:', sizes[0] if sizes else 0)
    print('Top 5 component sizes:', sizes[:5])
```

## D4. Node attributes: labels and sensitive attribute

```python
def get_node_attr(G, attr):
    return [G.nodes[n].get(attr) for n in G.nodes() if attr in G.nodes[n]]

sens = get_node_attr(G, 'sensitive')
labels = get_node_attr(G, 'label')

if sens:
    print('Sensitive attribute value counts:')
    print(pd.Series(sens).value_counts().sort_index())
if labels:
    print('\nLabel value counts:')
    print(pd.Series(labels).value_counts().sort_index())

if sens and labels and len(sens) == len(labels):
    cross = pd.crosstab(pd.Series(sens), pd.Series(labels))
    print('\nCross-tab (sensitive x label):')
    print(cross)
    print('→ Use for initial fairness view (e.g., demographic parity).')
```

## D5. Centrality (course alignment): PageRank or degree

```python
if G.number_of_nodes() > 0 and G.number_of_edges() > 0:
    try:
        pr = nx.pagerank(G)
    except Exception:
        pr = dict(G.degree())  # fallback to degree
    top = sorted(pr.items(), key=lambda x: x[1], reverse=True)[:10]
    print('Top-10 nodes by PageRank (or degree):')
    for node, score in top:
        s = G.nodes[node].get('sensitive', 'N/A')
        print(f'  {node}: score={score:.4f}, sensitive={s}')
else:
    print('Graph too small or empty for centrality.')
```

## D6. Bias considerations

```python
print('Bias notes:')
print('- Label imbalance can lead to biased node classification; report class distribution.')
print('- Sensitive-attribute imbalance (e.g., one group rare) affects fairness metrics (demographic parity, equalized odds).')
print('- Homophily: if edges connect same sensitive group more often, structure can reinforce bias.')
```
